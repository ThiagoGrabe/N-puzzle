```python
import matplotlib.pyplot as plt

def plot_result(algorithm, time, explored, visited):

    f, axarr = plt.subplots(3, sharex=True, sharey=True)
    f.suptitle('Results for Algorithm')
    plt.xlabel('x label')
    
    axarr[0].plot(time, color='g')
    axarr[1].plot(explored, color='b')
    axarr[2].plot(visited, color='r')
    # Bring subplots close to each other.
    f.subplots_adjust(hspace=0)
    # Hide x labels and tick labels for all but bottom plot.
    for ax in axarr:
#         plt.ylabel('y label')
        ax.label_outer()

    
    
    
def plot_result_2(algorithm, time, explored, visited):
    x1 = time
    x2 = explored
    x3 = visited
    
    plt.figure(figsize=(20,10))
    

    plt.subplot(3, 1, 1)
    plt.plot(x1, 'o-')
    plt.title('A tale of 2 subplots')
    plt.ylabel('Time')

    plt.subplot(3, 1, 2)
    plt.plot(x2,'.-')
    plt.xlabel('time (s)')
    plt.ylabel('Explored')
    
    plt.subplot(3, 1, 3)
    plt.plot(x3, '.-')
    plt.xlabel('time (s)')
    plt.ylabel('Visited')

    plt.show()
```

```python
import random

algorithm = 'Name'
time = random.sample(range(0, 32), 10)
explored = random.sample(range(0, 32), 10)
visited = random.sample(range(0, 32), 10)

print(algorithm)
```

```python
plot_result_2(algorithm, time, explored, visited)
```
